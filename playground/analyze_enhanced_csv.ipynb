```python

import matplotlib.pyplot as plt

def object_parts_scatter2D(df, object_parts, color_col):

    # Number of rows and columns for subplots
    nrows = 1
    ncols = 3 

    # Create figure and axes
    fig, axs = plt.subplots(nrows, ncols, figsize=(15, 5))

    axs[0].scatter(df[object_parts[0]], df[object_parts[1]], c=df[color_col].astype('category').cat.codes)
    axs[0].set_xlabel(object_parts[0])  
    axs[0].set_ylabel(object_parts[1])
    axs[0].set_title(f"{object_parts[0]} vs {object_parts[1]}") 

    axs[1].scatter(df[object_parts[1]], df[object_parts[2]], c=df[color_col].astype('category').cat.codes)
    axs[1].set_xlabel(object_parts[1])  
    axs[1].set_ylabel(object_parts[2])
    axs[1].set_title(f"{object_parts[1]} vs {object_parts[2]}") 

    axs[2].scatter(df[object_parts[0]], df[object_parts[2]], c=df[color_col].astype('category').cat.codes)
    axs[2].set_xlabel(object_parts[0])
    axs[2].set_ylabel(object_parts[2])
    axs[2].set_title(f"{object_parts[0]} vs {object_parts[2]}")

    # Add color legend
    handles, labels = axs[0].get_legend_handles_labels()
    fig.legend(handles, labels, loc='upper right')

    # Adjust layout  
    fig.tight_layout()

    plt.show()
```

```python
# Create 3D scatter plot

def object_parts_scatter3D(df, object_parts, color_col):
    
    # Create figure and axes
    fig = plt.figure(figsize=(15, 5))
    ax = fig.add_subplot(111, projection='3d')

    ax.scatter(df[object_parts[0]], df[object_parts[1]], df[object_parts[2]], c=df[color_col].astype('category').cat.codes)
    ax.set_xlabel(object_parts[0])  
    ax.set_ylabel(object_parts[1])
    ax.set_zlabel(object_parts[2])
    ax.set_title(f"{object_parts[0]} vs {object_parts[1]} vs {object_parts[2]}") 

    plt.show()
```

```python
import matplotlib.patches as mpatches

def object_parts_scatter2D(df, object_parts, color_col):
    # Number of rows and columns for subplots
    nrows = 1
    ncols = 3 

    # Create figure and axes
    fig, axs = plt.subplots(nrows, ncols, figsize=(15, 5))
    
    # Get unique categories and corresponding colors
    categories = df[color_col].astype('category').cat.categories
    color_codes = df[color_col].astype('category').cat.codes
    unique_colors = [plt.cm.viridis(i) for i in np.linspace(0, 1, len(categories))]

    axs[0].scatter(df[object_parts[0]], df[object_parts[1]], c=color_codes, cmap='viridis')
    axs[0].set_xlabel(object_parts[0])  
    axs[0].set_ylabel(object_parts[1])
    axs[0].set_title(f"{object_parts[0]} vs {object_parts[1]}") 

    axs[1].scatter(df[object_parts[1]], df[object_parts[2]], c=color_codes, cmap='viridis')
    axs[1].set_xlabel(object_parts[1])  
    axs[1].set_ylabel(object_parts[2])
    axs[1].set_title(f"{object_parts[1]} vs {object_parts[2]}") 

    axs[2].scatter(df[object_parts[0]], df[object_parts[2]], c=color_codes, cmap='viridis')
    axs[2].set_xlabel(object_parts[0])
    axs[2].set_ylabel(object_parts[2])
    axs[2].set_title(f"{object_parts[0]} vs {object_parts[2]}")

    # Create custom legend
    # Create custom legend
    legend_patches = [mpatches.Patch(color=color, label=cat) for color, cat in zip(unique_colors, categories)]
    fig.legend(handles=legend_patches, loc='upper center', title=color_col, bbox_to_anchor=(0.5, 1.15), ncol=len(categories))

    # Adjust layout  
    fig.tight_layout()

    plt.show()

```

```python
object_parts = ["dog", "bicycle", "umbrella"]
object_parts_scatter2D(df_enhanced, object_parts, "Species")
```

```python
object_parts = ["head_total", "thorax_total", "abdomen_total"]
object_parts_scatter2D(df_enhanced, object_parts, "Species")
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

#remove categorical columns
df_plot = df_enhanced.select_dtypes(exclude=['object', 'category'])

# 1. Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_plot)  # 'data' is your dataset

# 2. Initialize PCA and fit
pca = PCA(n_components=2)  # Reduce data to 2 principal components as an example
principal_components = pca.fit_transform(scaled_data)

# 'principal_components' now contains the data transformed into the first two principal components
```

```python
# Convert the principal components to a DataFrame for visualization
df_pca = pd.DataFrame(data=principal_components, columns=['Principal Component 1', 'Principal Component 2'])
df_pca['Species'] = df_enhanced['Species'].copy()

# Plotting the PCA results
plt.figure(figsize=(10, 7))
sns.scatterplot(data=df_pca, x='Principal Component 1', y='Principal Component 2', hue='Species', palette='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Dataset')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```
